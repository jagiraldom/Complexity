```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
import numpy as np
```

### Part 1
Consider de following functions:
- $TA(n) = 1000\cdot n $
- $TB(n) = 200\cdot n\cdot log(n)$
- $TC(n) = 10 \cdot n^{2}$
- $TD(n) = 2^{n}$

#### TA(n)
This is a function of linear order $O(n)$

```python
x = np.linspace(1,100,100)
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot()
plt.plot(x,x*1000, 'r')
```

#### TB(n)
This is a function with complexity $O(n\cdot log(n))$

```python
x = np.linspace(1,100,100)
log = np.vectorize(math.log)
y = 200*x*log(x)
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot()
plt.plot(x,y, 'r')
```

#### TC(n)
This is a function with cuadratic order $O(n^2)$

```python
x = np.linspace(1,100,100)
y = 10*x**2
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot()
plt.plot(x,y, 'r')
```

#### TD(n)
This is a function with exponential order of growth $O(2^n)$

```python
x = np.linspace(1,15,100)
exp2 = np.vectorize(math.exp2)
y = exp2(x)
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot()
plt.plot(x,y, 'r')
```

```python
x = np.linspace(1,200,100)
x_exp = np.linspace(1,18,100)
log = np.vectorize(math.log)
exp2 = np.vectorize(math.exp2)
fig = plt.figure(figsize=(8,3))
ax = fig.add_subplot()
plt.plot(x, 1000*x, 'r')
plt.plot(x, 200*x*log(x), 'b')
plt.plot(x, 10*x**2, 'g')
```

Based on the classification of growth complexity and the graph of the functions we can say the following:
- For inputs that are not so large (<100) the TA, TB, TC algorithms have similar performance. However the asymptotic behavior of the linear function is the best.
The function with exponential complexity is out of consideration because of its high cost even for small inputs.
